## Simple Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
```

```python
# create a figure and an axes
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x)); # creates the figure and axes for you
```

```python
plt.plot(x, np.sin(x)); # creates the figure and axes for you
plt.plot(x, np.cos(x)); # creates the figure and axes for you
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
